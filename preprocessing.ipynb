```python
import pandas as pd
import numpy as np
import nltk 
from nltk import word_tokenize,sent_tokenize
from nltk.corpus import stopwords
import re #to remove urls
import  string #to remove punctuation
#spelling correction
from textblob import TextBlob
from nltk.stem.porter import PorterStemmer
```

```python
df=pd.read_csv('cleaned_tmdb_movies.csv')
df.head(1)
```

```python
#lower case conversion
df['overview']=df['overview'].str.lower()
```

```python
df.drop(columns=['genre_ids'],inplace=True)
df.head(1)
```

```python
df['overview']
```

```python
df['overview'][0]
```

```python
nltk.download('punkt')
nltk.download('stopwords')
stop_words = stopwords.words('english')
#stop_words
```

```python
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stop_words:
            new_text.append('')
        else:
            new_text.append(word)
    x=new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
df['overview']=df['overview'].astype(str).apply(lambda x:remove_stopwords(x))
```

```python
print(string.punctuation)
exclude=string.punctuation
exclude+='“”‘’—…¬'
print(exclude)

def remove_punc(text):
  for char in exclude:
    text=text.replace(char,'')
  return text
```

```python
df['overview']=df['overview'].astype(str).apply(lambda x:remove_punc(x))
df['overview']
```

```python
#tokenization
import spacy
nlp =spacy.load('en_core_web_sm')

def word_tokenize_spacy(text):
    doc = nlp(text)
    return [token.text for token in doc]
```

```python
df['overview']=df['overview'].apply(word_tokenize_spacy)
df['overview']
```

```python
df['overview'] = df['overview'].apply(
    lambda tokens: [t for t in tokens if t.isalpha()]
)
df['overview']
```

```python
df['overview'] = df['overview'].apply(
    lambda tokens: [t.strip() for t in tokens if t.strip()]
)
df['overview']
```

```python
#lemmatization/stemming
'''
1) Porter Stemmer :nltk
2) snowball stemmer (improved proter/porter2) :nltk
3) Lancaster Stemmer :nltk
4) Regex Stemmer :nltk
5)Lovins Stemmer :nltk

'''
'''
WordNet Lemmatizer1: nltk
spaCy Lemmatizer : spacy
TextBlob Lemmatizer :textblob
Pattern Lemmatizer :pattern
Stanford CoreNLP Lemmatizer :stanfordnlp
Gensim Lemmatization (Wrapper) :gensim

'''

ps=PorterStemmer()

def stem_tokens(tokens):
  return [ps.stem(t) for t in tokens]

df['overview'] = df['overview'].apply(stem_tokens)
df['overview']
```
