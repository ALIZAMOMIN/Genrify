```python
import requests
import pandas as pd

# ------------------ GENRE API ------------------

genre_url = "https://api.themoviedb.org/3/genre/movie/list?language=en"

headers = {
    "accept": "application/json",
    "Authorization": "Bearer eyJhbGciOiJIUzI1NiJ9.eyJhdWQiOiI4YjA0ZWY0ZjNkNDZjN2Q1NjE1ODJlOTJmZWZkM2NkMiIsIm5iZiI6MTc0ODkyMDU4OS42NTEsInN1YiI6IjY4M2U2OTBkYWE4NTRjY2Q0YTUzODdhZCIsInNjb3BlcyI6WyJhcGlfcmVhZCJdLCJ2ZXJzaW9uIjoxfQ.O7udcfgo4pp4lZoWWzdUd720HGY6-kNWgGeEP9--4Kc"
}

response = requests.get(genre_url, headers=headers)

# Extract list of genres
genres = response.json()["genres"]

# Create mapping: {id: name}
genre_map = {genre["id"]: genre["name"] for genre in genres}

print("GENRE MAP:", genre_map)
print("\n")

# ------------------ MOVIE API (MULTIPLE PAGES) ------------------

titles = []
overviews = []
genre_ids_all = []
genre_names_all = []

# Fetching multiple pages (1 to 5 for example)
for page in range(1, 10):

    print("Fetching Page:", page)

    movie_url = f"https://api.themoviedb.org/3/discover/movie?include_adult=false&include_video=false&language=en-US&page={page}&sort_by=popularity.desc"

    response = requests.get(movie_url, headers=headers)
    data = response.json()

    for movie in data["results"]:

        title = movie["title"]
        overview = movie["overview"]
        genre_ids = movie["genre_ids"]   # e.g. [28, 12]

        # Convert all IDs → names using genre_map
        genre_names = [genre_map[g] for g in genre_ids]

        titles.append(title)
        overviews.append(overview)
        genre_ids_all.append(genre_ids)
        genre_names_all.append(genre_names)

# ------------------ CREATE DATAFRAME ------------------

df = pd.DataFrame({
    "title": titles,
    "overview": overviews,
    "genre_ids": genre_ids_all,
    "genre_names": genre_names_all
})

print(df.head())
```

```python
csv_path = "./tmdb_movies.csv"
df.to_csv(csv_path, index=False)

print("Saved to:", csv_path)
```
